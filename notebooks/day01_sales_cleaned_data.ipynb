```python
df['Order_Date'] = pd.to_datetime(
    df['Order_Date'],
    errors='coerce')
df.info()
```

```python
df['Order_Date']
```

I changed the 'Order Date' column to datetime dtype.

```python
print(df['Order_Date'].isna().sum())
```

All fields are correct.

```python
df['year'] = df['Order_Date'].dt.year
df['month'] = df['Order_Date'].dt.month
df['day'] = df['Order_Date'].dt.day
df['weekday_1to7'] = df['Order_Date'].dt.isocalendar().day
df['quarter'] = df['Order_Date'].dt.quarter
```

```python
df.weekday_1to7
```

```python
df['weekday_name'] = df['Order_Date'].dt.day_name()  # for plots
df['weekday_name']
```

```python
df['is_weekend'] = (df['weekday_1to7'] >= 6).astype(int)
df['is_weekend']
```

I created the additional columns for better Exloratory Data Analysis.

```python
df['Postal_Code'] = df['Postal_Code'].astype("object")
```

```python
df.isnull().sum()
```

```python
df['Region'] = df['Region'].str.lower().str.strip()
df['Category'] = df['Category'].str.lower().str.strip()
```
